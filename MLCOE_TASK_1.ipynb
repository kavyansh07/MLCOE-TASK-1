**ANALYZATION OF THE DATASET**

```python
color=data['Gender'].value_counts().plot(kind='pie')
```

There are more Male customers as compared to Female customers.

```python
data['Item Purchased'].value_counts().plot(kind='bar')
```

This bar graph shows that Bluse, Jwelley and Pants are the top 3 items purchased from the store whereas Backpack, Gloves, Jeans are the less purchased item of the store.


```python
color=data['Color'].value_counts().plot(kind='pie')
```

```python
sns.violinplot(x='Category', y='Purchase Amount (USD)', hue='Gender', data=data, split=True)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x='Location', y='Purchase Amount (USD)', data=data)
plt.xticks(rotation=90)
```

Maximum purchase are made from Anzona, Alaska, Pennsylvania, Washington and Michigan
